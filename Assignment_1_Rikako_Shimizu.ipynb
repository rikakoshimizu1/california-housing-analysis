#Task 3: Data Filtering

#### Points: 1


````python
"""
- Create a new DataFrame called 'filtered_housing' containing only the rows where 'total_rooms' are greater than 100 and housing_median_age is less than 5.

1 points
"""
filtered_housing=None

#Start code

#Placed rows where total_rooms > 100 and housing_median_age < 5 in filtered_housing dataFrame
filtered_housing = california_housing_dataframe[(california_housing_dataframe['total_rooms'] > 100) & (california_housing_dataframe['housing_median_age'] < 5)]

#End code

#Total number of rows in the new dataframe
print("Number of rows:", len(filtered_housing))
# Print the first few rows of the filtered DataFrame
filtered_housing.head()
````

````python
"""
- Normalize median_income using NumPy functions like np.mean() and np.std() and add column 'median_income_normalized' into the dataframe

1 points
"""
#Start code

#Import NumPy library
import numpy as np

#Convert median_income Pandas dataFrame into a NumPy array
np_array = california_housing_dataframe['median_income'].to_numpy()

#Calculate the mean of median_income
mean_income = np.mean(np_array)

#Calculate the standard deviation of median_income
sd_income = np.std(np_array)

#Create an empty NumPy array
median_income_normalized = np.array([])

#Calculate the normalized median_income
for i in california_housing_dataframe['median_income']:
  median_income_normalized_i = (i - mean_income)/sd_income
  median_income_normalized = np.append(median_income_normalized, median_income_normalized_i)

#Add median_income_normalized into the dataFrame
california_housing_dataframe['median_income_normalized'] = median_income_normalized.tolist()

#End code

california_housing_dataframe.head()
````

#Task 4: Data Transformation:

#### Points: 2

````python
"""
- Normalize total_rooms using NumPy functions like np.mean() and np.std() and add column 'total_rooms_normalized' to the dataframe

1 points
"""
#Start code

#Import NumPy library
import numpy as np

#Convert total_rooms Pandas dataFrame into a NumPy array
np_total_rooms = california_housing_dataframe['total_rooms'].to_numpy()

#Calculate the mean of total_rooms
mean_rooms = np.mean(np_total_rooms)

#Calculate the standard deviation of total_rooms
sd_rooms = np.std(np_total_rooms)

#Create an empty NumPy array
total_rooms_normalized = np.array([])

#Calculate the normalized total_rooms
for i in california_housing_dataframe['total_rooms']:
  total_rooms_normalized_i = (i - mean_rooms)/sd_rooms
  total_rooms_normalized = np.append(total_rooms_normalized, total_rooms_normalized_i)

#Add total_rooms_normalized into the dataFrame
california_housing_dataframe['total_rooms_normalized'] = total_rooms_normalized.tolist()

#End code

california_housing_dataframe.head()
````

#Task 5: Create a scatter plot of median_income_normalized vs. median_house_value

####Points: 2

````python
"""
- Creating scatter plots of median_income_normalized vs. median_house_value
2 points
"""

#Start code

#Plot the data points with 'x' points in blue
x = california_housing_dataframe['median_income_normalized']
y = california_housing_dataframe['median_house_value']
plt.scatter(x, y, marker='x', c='b')

#Set the title
plt.title("Median Income Normalized vs. Median House Value")

#Set the x-axis label
plt.xlabel('Normalized Median Income')

#Set the y-axis label
plt.ylabel('Median House Value (in dollars)')

plt.show()

#End code
````

#Task 6: Fit a linear regression model

Points: 1 point

````python
"""
- Create a linear regression model as 'reg' variable by calling LinearRegression from sklearn library imported
- Fit the model

1 points
"""
#importing the LinearRegression model from sklearn
from sklearn.linear_model import LinearRegression

#Defining X and y for training the dataset
#X contains the columns names (features) we want use and y the value we are trying to predict
X=california_housing_dataframe[['households', 'median_income_normalized', 'total_rooms_normalized','housing_median_age','population','longitude','latitude']]
y=california_housing_dataframe['median_house_value']

#Start code

#Linear regression model as 'reg' variable by calling LinearRegression from sklearn
#Fit the model
reg = LinearRegression().fit(X,y)
print(reg.coef_, reg.intercept_)

#End code
y_pred = reg.predict(X)
````

````python
#Let us check the model performance using MSE and R-squared as the metrics
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
r2_score = reg.score(X, y)
print(f"R-squared (R2) Score: {r2_score}")
````
